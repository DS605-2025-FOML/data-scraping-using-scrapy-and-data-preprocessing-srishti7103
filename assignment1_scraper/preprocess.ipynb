```python
import pandas as pd
data_books =pd.read_csv("C:/Users/srrml/Desktop/ML assignments/Lab02-srishti7103/data-scraping-using-scrapy-and-data-preprocessing-srishti7103/assignment1_scraper/books.csv")
print("Dataset books:\n",data_books.head())

data_books.rename(columns={
    'available_stocks': 'Books Available',
    'image_url': 'Image URL',
    'price': 'Price',
    'rating': 'Rating',
    'title': 'Title',
}, inplace=True)

print("Dataset books:\n",data_books.head())

data_books.drop(columns = ['Books Available'], inplace=True)

data_books.head()

data_books = data_books[['Title','Price','Rating','Image URL']]

data_books.head()

print("Dataset Information :\n",data_books.info)

print("Dataset stats :\n", data_books.describe())

data_books.head()

data_books.isnull()

data_books.isna().sum()

data_books.columns = data_books.columns.str.strip()
data_books = data_books.applymap(lambda x: x.strip() if isinstance(x, str) else x)

data_books['Price'] = data_books['Price'].str.replace('£', '', regex=False).astype(float)

rating_map = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}
data_books['Rating'] = data_books['Rating'].map(rating_map)
data_books.head()
data_books.describe()
data_books.describe

data_books.drop_duplicates(inplace=True)
data_books.dropna(inplace=True)
data_books.reset_index(drop=True, inplace=True)
data_books.head()
data_books.describe()
data_books.isnull().sum()
data_books.info

data_books.to_csv("books_cleaned_data.csv", index=False)

print("Books Shape:", data_books.shape)
```
